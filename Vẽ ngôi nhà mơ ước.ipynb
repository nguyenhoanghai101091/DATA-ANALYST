```python
#Vẽ ngôi nhà mơ ước
import turtle
a=turtle.Turtle()
a.speed(10)
#Vẽ hình tam giác tím
a.fillcolor("purple")
a.begin_fill()
a.left(120)
a.forward(100)
a.left(120)
a.forward(100)
a.left(120)
a.forward(100)
a.end_fill()
#Vẽ hình chữ nhật xanh dương
a.fillcolor("blue")
a.begin_fill()
a.right(90)
a.forward(100)
a.right(90)
a.forward(100)
a.right(90)
a.forward(100)
a.end_fill()
#vẽ hình chữ nhật màu xanh là cây
a.backward(100)
a.right(90)
a.penup()
a.forward(25)
a.fillcolor("green")
a.begin_fill()
a.pendown()
a.left(90)
a.forward(50)
a.right(90)
a.forward(50)
a.right(90)
a.forward(50)
a.end_fill()
#Vẽ hình bình hành vàng
a.left(90)
a.forward(25)
a.fillcolor("yellow")
a.begin_fill()
a.left(30)
a.forward(100)
a.left(60)
a.forward(100)
a.left(120)
a.forward(100)
a.end_fill()
#Vẽ hình bình hành màu nâu
a.penup()
a.backward(25)
a.left(60)
a.forward(25)
a.fillcolor("brown")
a.begin_fill()
a.pendown()
a.forward(25)
a.left(120)
a.forward(50)
a.left(60)
a.forward(25)
a.left(120)
a.forward(50)
a.end_fill()
#Về vị trí ban đầu
a.penup()
a.goto(0,0)
#vẽ hình bình hành màu cam
a.fillcolor("orange")
a.begin_fill()
a.pendown()
a.right(90)
a.forward(100)
a.right(90)
a.forward(100)
a.right(90)
a.forward(100)
a.end_fill()
#di chuyển đến điểm mốc vẽ cây
a.penup()
a.goto(0,0)
a.left(60)
a.goto(200,100)
a.pendown()
#bắt đầu vẽ cây xanh
#Vẽ tam giác màu xanh thứ 1
a.fillcolor("green")
a.begin_fill()
a.right(60)
a.forward(100)
a.right(120)
a.forward(100)
a.right(120)
a.forward(100)
a.end_fill()
#Di chuyển xuống vị trí đỉnh tam giác 2
a.penup()
a.right(120)
a.forward(100)
a.right(120)
a.forward(50)
a.right(180)
#Vẽ tam giác màu xanh thứ 2
a.pendown()
a.fillcolor("green")
a.begin_fill()
a.right(60)
a.forward(120)
a.right(120)
a.forward(120)
a.right(120)
a.forward(120)
a.end_fill()
#Di chuyển xuống vị trí đỉnh tam giác 2
a.penup()
a.right(120)
a.forward(120)
a.right(120)
a.forward(60)
a.right(180)
#Vẽ tam giác màu xanh thứ 3
a.pendown()
a.fillcolor("green")
a.begin_fill()
a.right(60)
a.forward(140)
a.right(120)
a.forward(140)
a.right(120)
a.forward(140)
a.end_fill()
#Di chuyển xuống vị trí bắt đầu vẽ hình chữ nhật màu nâu
a.penup()
a.right(120)
a.forward(140)
a.right(120)
a.forward(50)
#vẽ hình chữ nhật màu nâu
a.pendown()
a.fillcolor("brown")
a.begin_fill()
a.left(90)
a.forward(100)
a.right(90)
a.forward(40)
a.right(90)
a.forward(100)
a.end_fill()
#Di chuyển rùa đến điểm mốc vẽ mặt trời
a.penup()
a.right(90)
a.goto(-200,200)
a.pencolor("black")
a.pendown()
#Vẽ mặt trời
a.forward(120)
a.backward(240)
a.goto(-200,200)
a.left(45)
a.forward(100)
a.backward(200)
a.goto(-200,200)
a.left(45)
a.forward(120)
a.backward(240)
a.goto(-200,200)
a.left(45)
a.forward(100)
a.backward(200)
a.goto(-200,200)
a.pencolor("yellow")
a.dot(150)
a.pencolor("black")
a.penup()
a.backward(75)
a.right(90)
a.pendown()
a.circle(75)
turtle.done()
```
