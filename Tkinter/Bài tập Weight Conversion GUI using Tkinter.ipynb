```python
#Weight Conversion GUI using Tkinter
import tkinter as tk
from tkinter import ttk

#Khai báo hàm chuyển đổi

def callback():
    gram_val = float(kg_input.get())*1000
    pounds_val = float(kg_input.get())*2.20462
    ounce_val = float(kg_input.get())*35.274 
#output change
    gram_result.config(text = gram_val)
    pounds_result.config(text = pounds_val)
    ounce_result.config(text = ounce_val)

#tạo cửa sổ

main = tk.Tk()
main.geometry("500x150")
main.title("Weight Conversion")

#row1
kg_input_label = ttk.Label(main, text="Enter the weight in Kg")
kg_input = ttk.Entry(main)
convert_btn = ttk.Button(main, text="Convert", command = callback)
kg_input_label.grid(row=0, column=0)
kg_input.grid(row=0, column=1)
convert_btn.grid(row=0, column=2)
#row2
gram_label = ttk.Label(main, text="Gram")
pounds_label = ttk.Label(main, text="Pounds")
ounce_label = ttk.Label(main, text="Ounce")
gram_label.grid(row=1, column=0)
pounds_label.grid(row=1, column=1)
ounce_label.grid(row=1, column=2)
#row3
gram_result = ttk.Label(main, text="0")
pounds_result = ttk.Label(main, text="0")
ounce_result = ttk.Label(main, text="0")
gram_result.grid(row=2, column=0)
pounds_result.grid(row=2, column=1)
ounce_result.grid(row=2, column=2)
main.mainloop()
```
