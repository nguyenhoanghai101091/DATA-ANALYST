```python
import turtle
t=turtle.Turtle()
#Vẽ 2 hình vuông đồng tâm
t.penup()
t.forward(100)
t.left(135)
t.pendown()
t.forward(100)
t.left(90)
t.forward(100)
t.left(90)
t.forward(100)
t.left(90)
t.forward(100)
t.left(90)
t.penup()
t.goto(25,0)
t.backward(75)
t.left(45)
t.pendown()
t.forward(100)
t.right(90)
t.forward(100)
t.right(90)
t.forward(100)
t.right(90)
t.forward(100)
t.right(90)
turtle.done()
```
