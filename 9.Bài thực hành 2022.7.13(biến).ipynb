```python
#bài 3
import turtle
import math
t=turtle.Turtle()
green=turtle.Screen()
green.bgcolor('green')
t.speed(10)
#Khai báo cạnh hình vuông
canh=100
khoangcach=40
cos45=math.cos(math.radians(45))
t.penup()
t.goto(-canh/2,-canh/2)
t.pendown()
#vẽ hình vuông 1
t.pencolor('red')
t.forward(canh)
t.left(90)
t.forward(canh)
t.left(90)
t.forward(canh)
t.left(90)
t.forward(canh)
#rẽ hướng sang hình thứ 2
t.penup()
t.right(45)
t.forward(khoangcach/cos45)
t.left(135)
#vẽ hình vuông 2
t.pendown()
t.pencolor('yellow')
t.forward(khoangcach*2+canh)
t.left(90)
t.forward(khoangcach*2+canh)
t.left(90)
t.forward(khoangcach*2+canh)
t.left(90)
t.forward(khoangcach*2+canh)
#rẽ hướng sang hình thứ 3
t.penup()
t.right(45)
t.forward(khoangcach/cos45)
t.left(135)
#vẽ hình vuông 3
t.pendown()
t.pencolor('blue')
t.forward(khoangcach*4+canh)
t.left(90)
t.forward(khoangcach*4+canh)
t.left(90)
t.forward(khoangcach*4+canh)
t.left(90)
t.forward(khoangcach*4+canh)
turtle.done()
```
